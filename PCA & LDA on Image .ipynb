```python
#Importing required Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# Load the Olivetti Faces dataset
faces = fetch_olivetti_faces()
X = faces.data
y = faces.target
```

```python
# Compute the PCA and LDA projections
pca = PCA(n_components=100)
X_pca = pca.fit_transform(X)
lda = LinearDiscriminantAnalysis(n_components=39)
X_lda = lda.fit_transform(X, y)
```

```python
# Plot the PCA and LDA projections
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, alpha=0.8, cmap='jet')
plt.title('Principal Component Analysis')
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,2)
plt.scatter(X_lda[:,0], X_lda[:,1], c=y, alpha=0.8, cmap='jet')
plt.title('Linear Discriminant Analyis')
plt.show()
```
